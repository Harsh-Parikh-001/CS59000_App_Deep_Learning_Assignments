
Remove the blank rows

```python
new_data = data.dropna()
new_data_1 =  len(new_data["Extra"])
print(new_data_1)
```

```python
co_coefficiant = new_data["Extra"].corr(new_data["Total"])
print(co_coefficiant)
```

```python
plt.scatter(new_data["Extra"], new_data["Total"])
plt.xlabel("extra credits")

plt.ylabel("final score")
plt.show()
```

```python
X = new_data['Extra']
Y = new_data['Total']
X_train, X_valid, y_train, y_valid = train_test_split(X, Y, train_size=0.8, test_size=0.2,random_state = 420)
X = np.array(X_train).reshape(-1,1)
y = np.array(y_train).reshape(-1,1)
X_valid = np.array(X_valid).reshape(-1,1)
model = LinearRegression()
model.fit(X, y)

predictedVals = model.predict(X_valid)
```

```python
mse = mean_squared_error(y_valid, predictedVals)
RMSE = math.sqrt(mse)
print(RMSE)
```
